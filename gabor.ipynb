```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
size=15
theta=1*np.pi/2
sigma=5
lamda=1*np.pi/4
gamma=0.9
phi=0.7

kernel=cv2.getGaborKernel((size, size), sigma, theta, lamda, gamma, phi, ktype=cv2.CV_32F)
plt.imshow(kernel)
```
